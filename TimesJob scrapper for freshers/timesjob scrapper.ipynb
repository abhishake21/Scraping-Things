```python
def timesjobsearch(searchString,exp,numofjobsperpage=20,numofpages=5):
   
    try:
        file = open(searchString+'.csv','w',newline='')
        writer = csv.writer(file)
        writer.writerow(['Job Role','Company Name','Key Skills','Experience','Location','Job Posted','Job Link'])
        
        for page in range(1, numofpages+1):
            url = requests.get("https://www.timesjobs.com/candidate/job-search.html?from=submit&luceneResultSize="+str(numofjobsperpage)+"&cboWorkExp1="+str(exp)+"&postWeek=30&txtKeywords="+searchString+"&sequence="+str(page)).text
            soup = BeautifulSoup(url, 'html')
            jobs = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx')

            for job in jobs:
                try:
                    job_role = job.find('header', class_ = 'clearfix').h2.text.strip()
                except:
                    job_role = "Job Role not specified"

                try:
                    company_name = job.find('h3', class_ = 'joblist-comp-name').text.replace('(More Jobs)','').strip()
                except:
                    company_name = "Company Name not specified"

                try:
                    key_skills = job.find('span', class_ = 'srp-skills').text.replace(' ','').strip()
                except:
                    key_skills = "Skills not specified"

                try:
                    experience = job.find('ul', class_ = 'top-jd-dtl clearfix').li.text.replace('card_travel','')
                except:
                    experience = "Experience not specified"

                try:
                    location = job.find('ul', class_ = 'top-jd-dtl clearfix').span.text
                except:
                    location = 'Location not Specified'

                try:
                    job_posted = job.find('span', class_ = 'sim-posted').span.text.replace('Work From Home','').strip()
                except:
                    job_posted = "Not Specified"

                try:
                    job_link = job.header.h2.a['href']
                except:
                    job_link = "Link no specified"

                writer.writerow([job_role,company_name,key_skills,experience,location,job_posted,job_link])

        file.close()
        print(searchString+'.csv with '+str(exp)+' experience file created.')

    except:
        print('''Something went wrong.

Try reducing numofpages or numofjobsperpage(max value is 200).

If not specified, the default values are numofjobsperpage=10, numofpages=5 that will result in total 100 jobs.

If reducing the above values doesnt work then i dont know. Check the code yourself :)''')
```
