```python
from selenium import webdriver as wd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time
import wget
import os
from IPython.display import Image
```

```python
def instadown(searchString, num=3):
    
    driver = wd.Chrome('chromedriver.exe')
    driver.get('https://www.instagram.com/')

    usern = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='username']")))
    passwd = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='password']")))
    usern.clear()
    passwd.clear()

    usern.send_keys('Your-Username')
    passwd.send_keys('Your-Password')
    login = driver.find_element_by_xpath('//*[@id="loginForm"]/div/div[3]/button/div').click()

    not_now = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,"//button[contains(text(), 'Not Now')]"))).click()
    not_now2 = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,"//button[contains(text(), 'Not Now')]"))).click()
    
    searchbox = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH , "//*[@id='react-root']/section/nav/div[2]/div/div/div[2]/input")))
    searchbox.clear()
    searchbox.send_keys(str(searchString))
    searchbox.send_keys(Keys.ENTER)
    searchbox.send_keys(Keys.ENTER)
    time.sleep(2)
    searchbox.send_keys(Keys.ENTER)
    searchbox.send_keys(Keys.ENTER)
    time.sleep(2)

    for j in range(0, num):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

    images = driver.find_elements_by_tag_name('img')
    images = [image.get_attribute('src') for image in images]

    path = os.getcwd()
    path = os.path.join(path, str(searchString))
    os.mkdir(path)

    count = 0
    for image in images:
        save_as = os.path.join(path, str(searchString) + str(count) + '.jpg')
        wget.download(image, save_as)
        count+=1
```
